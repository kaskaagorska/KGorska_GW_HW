```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv_file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
df = pd.read_csv(csv_file)
df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
cols = [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
df2 = df.iloc[:,cols]
df2.head()
```

```python
df2.columns
```

```python
# Change reading 0 and 1 to No and Yes, respectively
df2['AttendedBootcamp'] = df2['AttendedBootcamp'].apply(lambda x:0 if x!= 1 else 1)
```

```python
# Extract rows for only those who attended a bootcamp
df_bootcamp = df2[df2['AttendedBootcamp'] == 1]
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
significant_df = pd.DataFrame()
for i in df_bootcamp.groupby(['BootcampName']):
    camp_df = i[1]
#     display(camp_df.head())
    count = camp_df.shape[0]
#     print(counts)
    if count > 20:
#         print(count)
        significant_df = significant_df.append(camp_df)
        
```

```python
significant_df = significant_df.reset_index().drop('index',axis=1)
```

```python
significant_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
for i in df_bootcamp[df_bootcamp['BootcampRecommend'] == 1].groupby(['BootcampName']):
    print("Bootcamp Name {}  Recommended Count {}".format(i[0],i[1].shape[0]))
```

```python
# Merge the two created data frames on the name of tbe bootcamp
pd.merge(df_bootcamp,significant_df,on='BootcampName')
```

```python
# Calculate percentage of each bootcamp's students who are recommenders

data = []
for i in df_bootcamp.groupby(["BootcampName"]):
    camp_df = i[1]
    total = camp_df.shape[0]
    rec = camp_df[camp_df["BootcampRecommend"] == 1].shape[0]
    pct = (rec/total)*100
    data.append([i[0],total,rec,pct])
    
```

```python
out_df = pd.DataFrame(data,columns=["BootcampName","Count","Recommenders","% Recommenders"])
```

```python
out_df.to_excel("output.xlsx")
```
